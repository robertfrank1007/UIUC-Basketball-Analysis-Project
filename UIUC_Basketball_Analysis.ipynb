## Section 1: Dataset

I chose to use a dataset containing the Illinois Fighting Illini basketball stats through their fitrst 6 games of the 2023-2024 season. College basketball, especially the Illini, is my favorite sport to watch, so I am curious to see how each player has done so far.

```python
import pandas as pd
df = pd.read_csv("IllinoisBaketball.csv")
df
```

```python
df_numeric = df.apply(pd.to_numeric, errors='coerce')
df_numeric
```

## Section 2: Exploratory Data Analysis

We can already see that there are 13 players on the team, but figuring out the means for each stat will help us see who is performing above or below the average. I also only included the columns with numerical values.

```python
df_numeric.mean()
```

## Section 3: Exploratory Data Visualization

```python
df["FG%"].plot.hist(bins = 30)
```

This is a histogram of the field goal percentages of the 13 players on the Illinois roster. This shows that 3 players hover around the 40% mark, and the data ranges from 0% all the way to 100%. The average for the team is 42% and this graph shows that there is a roughly equal numeber of players on either side of the average.

## Section 4: Data Science

How strong is the correlation between 3P% and points per game? Because 3 pointers are worth more, if you can shoot 3's well, this could potentially correlate to more points.

```python
df_numeric = df.apply(pd.to_numeric, errors='coerce')
df_numeric.corr()
```

The correlation coefficient is 0.510092 which indicates that there is medium positive correlation between 3P% and points per game.
